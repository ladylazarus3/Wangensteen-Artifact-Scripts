```python
# Import needed modules 

import pandas
import csv
import pymarc
import re
import math
from datetime import date
from pymarc import Record
```

```python
# Open CSV file with objects metadata for reading
with open("Wangensteen-objects-input.csv", encoding='utf-8') as infile:
    inread = csv.reader(infile)
    inlist = list(inread)
```

```python
# Create variables for dates and create *.mrc file

today = str(date.today())
today008 = today[2:].replace('-','')

processednew = open('new-wang-objects_' + today + '_out.mrc', 'wb')
processedupdate = open('update-wang-objects_' + today + '_out.mrc', 'wb')
```

```python
from csv import writer
def append_list_as_row(file_name, list_of_elem):
    # Open file in append mode
    with open('barcode-storageiddata' + today +  '_out.csv', 'a+', newline='') as write_obj:
        # Create a writer object from csv module
        csv_writer = writer(write_obj)
        # Add contents of list as last row in the csv file
        csv_writer.writerow(list_of_elem)
```

```python
for row in inlist[1:]:
    brief_record = pymarc.Record(to_unicode=True, force_utf8=True, leader = '     nrm a22     2i 4500')
    
    #Assign csv values
    status = ''
    status = row[1].strip()
    #photofilename = ''
    #photofilename = row[2].strip()
    #photourl = ''
    #photourl = row[3].strip()
    reviewpipeline = row[2].strip()
    barcodescanfield = row[3].strip()
    mmsid = ''
    mmsid = row[4].strip()
    datetype = ''
    datetype = row[5].strip()
    startdate = row[6].strip()
    enddate = row[7].strip()
    language = row[8].strip()
    tagnumber = ''
    tagnumber = row[9].strip()
    creator = ''
    creator = row[10].strip()
    objecttitle = ''
    objecttitle = row[11].strip()
    alttitle = []
    alttitle = str(row[12].strip())
    sourceoftitle = ''
    sourceoftitle = row[13]
    placeofmanufacture = ''
    placeofmanufacture = row[14].strip()
    nameofmanufacturer = ''
    nameofmanufacturer = row[15].strip()
    dateofmanufacture = ''
    dateofmanufacture = row[16].strip()
    objectextent = ''
    objectextent = row[17].lower().strip()
    objectmaterial = ''
    objectmaterial = row[18].lower().strip()
    heightcm = ''
    heightcm = row[19]
    roundedheight = ''
    widthcm = ''
    widthcm = row[20]
    roundedwidth = ''
    lengthcm = ''
    lengthcm = row[21]
    roundedlength = ''
    accompmat = ''
    accompmat = row[22]
    generalnote = ''
    generalnote = row[23]
    contents = ''
    contents = row[24]
    restrictions = ''
    restrictions = row[25]
    bibcitation = ''
    bibcitation = row[26]
    bibcitationloc = ''
    bibcitationloc = row[27]
    summaryscope = ''
    summaryscope = row[28]
    condition = ''
    condition = row[29]
    gift = ''
    gift = row[30]
    storageneeds = ''
    storageneeds = row[31].strip()
    subject1 = ''
    subject1 = row[32].strip()
    subject1q = ''
    subject1q = row[33].strip()
    subject2 = ''
    subject2 = row[34].strip()
    subject2q = ''
    subject2q = row[35].strip()
    collection = ''
    collection = row[36].strip()
    contribperson = ''
    contribperson = str(row[37].strip())
    corp1 = ''
    corp1 = str(row[38].strip())
    location = ''
    location = row[39].strip()
    warning = ''
    warning = row[40].strip()
    barcode = ''
    barcode = row[41].strip()
    storageID = row[42].strip()
    print(mmsid)
    #Split multiple 246s
    alttitle = alttitle.split("|")
    print(len(alttitle))
    print(alttitle) 
    print(bool(alttitle))
    
    #Split multiple 510s a & c
    bibcitation = bibcitation.split("|")
    bibcitationloc = bibcitationloc.split("|")
    
    #Split multiple 700s
    contribperson = contribperson.split("|")
    print(len(contribperson))
    print(contribperson)
    print(bool(contribperson))
    
    #Split multiple 710s
    corp1 = corp1.split("|")
    print(len(corp1))
    print(corp1)
    print(bool(corp1))
    
    #assign values 008
    data_008 = today008 + datetype + startdate + enddate + 'xx nnn            rn' + language + ' d'
    print(data_008) 
    
    #dealing with no extent
    
    if objectextent == '':
        objectextent = '1 object'

    #dealing with dimensions
    
    # Rounding dimensions as needed.
    heightnumbersonly = re.findall("\d\.\d|\d", heightcm)
    if len(heightnumbersonly) > 0:
        joinedheight = ''
        joinedheight = (joinedheight.join(heightnumbersonly))
        roundedheight = str(math.ceil(float(joinedheight)))
        print(roundedheight)
    else:
        roundedheight = ''
        
    widthnumbersonly = re.findall("\d\.\d|\d", widthcm)
    if len(widthnumbersonly) > 0:
        joinedwidth = ''
        joinedwidth = (joinedwidth.join(widthnumbersonly))
        roundedwidth = str(math.ceil(float(joinedwidth)))
        print(roundedwidth)
    else:
        roundedwidth = ''
        
    lengthnumbersonly = re.findall("\d\.\d|\d", lengthcm)
    if len(lengthnumbersonly) > 0:
        joinedlength = ''
        joinedlength = (joinedlength.join(lengthnumbersonly))
        roundedlength = str(math.ceil(float(joinedlength)))
        print(roundedlength)
    else:
        roundedlength = ''
        
    #check values and bracketed phrases for 264 variables
    
    if len(placeofmanufacture) > 0:
        placeofmanufacture = placeofmanufacture
    else:
        placeofmanufacture = '[Place of manufacture not identified]'
        
    if len(nameofmanufacturer) > 0:
        nameofmanufacturer = nameofmanufacturer
    else:
        nameofmanufacturer = '[Manufacturer not identified]'
        
    if len(dateofmanufacture) > 0:
        dateofmanufacture = dateofmanufacture
    else:
        dateofmanufacture = '[date of manufacture not identified]'
        
    #Build MARC fields from csv values and constant data. New field variable names must begin with 'field_'.
    
    if len(mmsid) > 0:
        field_001 = pymarc.Field(tag = '001', data= mmsid)
        print(field_001)
        brief_record.add_ordered_field(field_001)
    else:
        print('record is new')

    field_008 = pymarc.Field(tag='008', data=data_008)
    brief_record.add_ordered_field(field_008)
    print(field_008)

    field_040 = pymarc.Field(tag = '040', 
                            indicators = [' ',' '],
                            subfields = ['a', 'MnU', 'b', 'eng', 'e', 'rda', 'c', 'MnU'])
    brief_record.add_ordered_field(field_040)
    print(field_040)

    if len(tagnumber) > 0:
        field_099 = pymarc.Field(tag = '099', 
                            indicators = [' ',' '],
                            subfields = ['a', tagnumber])
        brief_record.add_ordered_field(field_099)
        print(field_099)
    
    if len(creator) > 0:
        field_100 = pymarc.Field(tag = '100',
                            indicators = ['1',' '],
                            subfields = ['a', creator])
        brief_record.add_ordered_field(field_100)
        print(field_100)

    field_245 = pymarc.Field(tag = '245',
                            indicators = ['0','0'],
                            subfields = ['a', objecttitle])
    brief_record.add_ordered_field(field_245)
    print(field_245)
    
    if alttitle[0] != '':
        for title in alttitle:
            field_246 = pymarc.Field(tag = '246',
                            indicators = ['3',' '],
                            subfields = ['a', str(title).strip()])
            brief_record.add_ordered_field(field_246)
            print(field_246)

    field_264 = pymarc.Field(tag = '264',
                            indicators = [' ','3'],
                            subfields = ['a', placeofmanufacture + ' :', 'b', nameofmanufacturer + ',', 'c', dateofmanufacture])
    brief_record.add_ordered_field(field_264)
    print(field_264)

    if len(roundedheight) > 0:
        field_300 = pymarc.Field(tag = '300',
                                indicators = [' ',' '],
                                subfields = ['a', objectextent + ' :', 'b', objectmaterial + ' ;', 'c', str(roundedheight) + ' x ' + str(roundedlength) + ' x ' + roundedwidth + ' cm'])
        print(field_300)
        brief_record.add_ordered_field(field_300)
   
    if len(accompmat) > 0:
        field_300 = pymarc.Field(tag = '300',
                                indicators = [' ',' '],
                                subfields = ['e', accompmat])
        print(field_300)
        brief_record.add_ordered_field(field_300)
    
    field_336 = pymarc.Field(tag = '336',
                            indicators = [' ',' '],
                            subfields = ['a', 'three-dimensional form', 'b', 'tdf', '2','rdacontent'])
    print(field_336)
    brief_record.add_ordered_field(field_336)

    field_337 = pymarc.Field(tag = '337',
                            indicators = [' ',' '],
                            subfields = ['a', 'other', 'b', 'x', '2','rdamedia'])
    brief_record.add_ordered_field(field_337)
    print(field_337)

    field_338 = pymarc.Field(tag = '338',
                            indicators = [' ',' '],
                            subfields = ['a', 'object', 'b', 'nr', '2','rdacarrier'])
    brief_record.add_ordered_field(field_338)
    print(field_338)
    
    if len(generalnote) > 0:
        field_500 = pymarc.Field(tag = '500',
                                indicators = [' ',' '],
                                subfields = ['a', generalnote])
        print(field_500)
        brief_record.add_ordered_field(field_500)
        
    if len(sourceoftitle) > 0:
        field_588 = pymarc.Field(tag = '588',
                                indicators = [' ',' '],
                                subfields = ['a', sourceoftitle])
        brief_record.add_ordered_field(field_588)
        print(field_588)
        
    if len(contents) > 0:
        field_505 = pymarc.Field(tag = '505',
                                indicators = ['0',' '],
                                subfields = ['a', contents])
        brief_record.add_ordered_field(field_505)
        print(field_505)
    else:
        print('no contents available')
        
        
    if len(restrictions) > 0:
        field_506 = pymarc.Field(tag = '506',
                                indicators = ['1',' '],
                                subfields = ['a', restrictions])
        brief_record.add_ordered_field(field_506)
        print(field_506)
    else:
        print('no summary scope available')
        
    if bibcitation != ['']:
        if bibcitationloc != ['']:            
            i = 0
            for citation in bibcitation:
                if i <= len(bibcitation):
                    page = bibcitationloc[i]
                    print("bibcitationloc length is =" + str(len(bibcitationloc)))
                    print("i length is =" + str(i))
                    print("page length is =" + str(len(page)))
                    i = i + 1
                    field_510 = pymarc.Field(tag = '510',
                                             indicators = ['4',' '],
                                             subfields = ['a', citation, 'c', page])
                    brief_record.add_ordered_field(field_510)
                else:
                    print("citation loc error handling")
            print(field_510)
        else:
            for citation in bibcitation:
                field_510 = pymarc.Field(tag = '510',
                                         indicators = ['0',' '],
                                         subfields = ['a', citation])
                brief_record.add_ordered_field(field_510)
            print(field_510)
    else:
        print('no bibcitation available')
    
    if len(summaryscope) > 0:
        field_520 = pymarc.Field(tag = '520',
                                indicators = [' ',' '],
                                subfields = ['a', summaryscope])
        brief_record.add_ordered_field(field_520)
        print(field_520)
    else:
        print('no summary scope available')
    
    if len(subject1) > 0:
        if len(subject1q) > 0:
            field_650 = pymarc.Field(tag = '650',
                                             indicators = [' ','2'],
                                             subfields = ['a', subject1, 'x', subject1q])
            brief_record.add_ordered_field(field_650)
            print(field_650)
        else:
            field_650 = pymarc.Field(tag = '650',
                                             indicators = [' ','2'],
                                             subfields = ['a', subject1])
            brief_record.add_ordered_field(field_650)
            print(field_650)
    else:
        print('no subject analysis 1 available')
    
    if len(subject2) > 0:
        if len(subject2q) > 0:
            field_650 = pymarc.Field(tag = '650',
                                             indicators = [' ','2'],
                                             subfields = ['a', subject2, 'x', subject2q])
            brief_record.add_ordered_field(field_650)
            print(field_650)
        else:
            field_650 = pymarc.Field(tag = '650',
                                             indicators = [' ','2'],
                                             subfields = ['a', subject2])
            brief_record.add_ordered_field(field_650)
            print(field_650)
    else:
        print('no subject analysis 2 available')

    if len(condition) > 0:
        field_866 = pymarc.Field(tag = '866',
                                indicators = [' ','1'],
                                subfields = ['z', condition + '.'])
        brief_record.add_ordered_field(field_866)
        print(field_866)
        
    if len(gift) > 0:
        field_866 = pymarc.Field(tag = '866',
                                indicators = [' ','1'],
                                subfields = ['z', gift + '.'])
        brief_record.add_ordered_field(field_866)
        print(field_866)
    
    if len(collection) > 0:
        field_710 = pymarc.Field(tag = '710',
                                indicators = ['2',' '],
                                subfields = ['a', collection + '.', '5', 'MnU'])
        brief_record.add_ordered_field(field_710)
        print(field_710)
        
    if contribperson[0] != '':
        for person in contribperson:
            field_700 = pymarc.Field(tag = '700',
                                    indicators = ['1',' '],
                                    subfields = ['a', str(person).strip()])
        brief_record.add_ordered_field(field_700)
        print(field_700)
        
    if corp1[0] != '':
        for corp in corp1:
            field_710 = pymarc.Field(tag = '710',
                                indicators = ['2',' '],
                                subfields = ['a', str(corp).strip()])
        brief_record.add_ordered_field(field_710)
        print(field_710)
    
    field_909 = pymarc.Field(tag = '909',
                            indicators = [' ',' '],
                            subfields = ['a', 'bcat', 'b', 'MNU', 'c', today, 'd','wang-objects'])
    print(field_909)
    brief_record.add_ordered_field(field_909)

    if len(storageneeds) > 0:
        if len(warning) > 0:
            field_981 = pymarc.Field(tag = '981',
                                     
                                    indicators = [' ',' '],
                                    subfields = ['c', 'TBWAN', 'd', location, 'e', barcode, 'j', tagnumber, 'm', warning, 't', 'p', 't', 'e', 'x', storageneeds])
            print(field_981)
            brief_record.add_ordered_field(field_981)
        else:
            field_981 = pymarc.Field(tag = '981',
                                    indicators = [' ',' '],
                                    subfields = ['c', 'TBWAN', 'd', location, 'e', barcode, 'j', tagnumber, 't', 'p', 't', 'e', 'x', storageneeds])
            print(field_981)
            brief_record.add_ordered_field(field_981)
    else:
        if len(warning) > 0:
            field_981 = pymarc.Field(tag = '981',
                                     
                                    indicators = [' ',' '],
                                    subfields = ['c', 'TBWAN', 'd', location, 'e', barcode, 'j', tagnumber, 'm', warning, 't', 'p', 't', 'e', 'x', storageneeds])
            print(field_981)
            brief_record.add_ordered_field(field_981)
        else:
            field_981 = pymarc.Field(tag = '981',
                                    indicators = [' ',' '],
                                    subfields = ['c', 'TBWAN', 'd', location, 'e', barcode, 'j', tagnumber, 't', 'p', 't', 'e'])
            brief_record.add_ordered_field(field_981)
        
    if len(mmsid) == 0:
        field_982 = pymarc.Field(tag = '982',
                            indicators = [' ',' '],
                            subfields = ['t', 'P'])
        print(field_982)
        brief_record.add_ordered_field(field_982)
    else:
        print('no physical items needed')
    
    #Add tags for update records which will only need electronic portfolios.
    
    field_982 = pymarc.Field(tag = '982',
                            indicators = [' ',' '],
                            subfields = ['t', 'E'])
    print(field_982)
    brief_record.add_ordered_field(field_982)    
    
    
    #Create list of all 'field_' variables and write to record

    if len(mmsid) > 0:
        processedupdate.write(brief_record.as_marc())
    else:
        processednew.write(brief_record.as_marc())
        
    #Write row to *.csv with barcode and Storage id
   
    row_contents = [barcode, storageID]
    append_list_as_row('barcode-storageiddata' + today + '_out.csv', row_contents)
    
    
    
    
```
