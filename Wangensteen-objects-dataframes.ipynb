```python
#Create variables in order
InputDate = str(date.today())
Action = 'output'
MMSID = ''
DateType = ''
StartDate = ''
EndDate = ''
Language = ''
TagNumber = ''
Creator = ''
Title = ''
AlternativeTitles = ''
SourceOfTitle = ''
PlaceManufacture = ''
NameManufacturer = ''
DateManufacurer = ''
ObjectExtent = ''
ObjectMaterials = ''
Height = ''
Width = ''
Length = ''
AccompanyingMaterial = ''
Contents = ''
RestrictionsMaterial = ''
BibCitation = ''
BibCitationLoc = ''
Description = ''
Condition = ''
GiftNotes = ''
StorageNeeds = ''
Mesh1 = ''
Mesh1Q = ''
Mesh2 = ''
Mesh2Q = ''
Collection = ''
ContributorInd = ''
ContributorCorp = ''
Location = ''
Barcode = ''
InventoryNo = ''
```

```python
#Write data from marc to MARCDF DataFrame

with open('Wangensteen-objects-dataframes-input.mrc', 'rb') as fh:
    reader = MARCReader(fh)
    for record in reader:
        #Create variables in order
        InputDate = str(date.today())
        Action = 'output'
        MMSID = ''
        DateType = ''
        StartDate = ''
        EndDate = ''
        Language = ''
        TagNumber = ''
        Creator = ''
        Title = ''
        AlternativeTitles = ''
        SourceOfTitle = ''
        PlaceManufacture = ''
        NameManufacturer = ''
        DateManufacurer = ''
        ObjectExtent = ''
        ObjectMaterials = ''
        Height = ''
        Width = ''
        Length = ''
        AccompanyingMaterial = ''
        Contents = ''
        RestrictionsMaterial = ''
        BibCitation = ''
        BibCitationLoc = ''
        Description = ''
        Condition = ''
        GiftNotes = ''
        StorageNeeds = ''
        Mesh1 = ''
        Mesh1Q = ''
        Mesh2 = ''
        Mesh2Q = ''
        Collection = ''
        ContributorInd = ''
        ContributorCorp = ''
        Location = ''
        Barcode = ''
        InventoryNo = ''
        
        #Grab MMS ID
        MMSID = str(record['001'])[6:]
        MMSIDnumber = str(MMSID)
        print(MMSIDnumber)
        
        #Grab DateType
        DateType = str(record['008'])[12]
        DateType = str(DateType)
        print(DateType)
        
        #Grab StartDate
        StartDate = str(record['008'])[13:17]
        StartDate = str(StartDate)
        print(StartDate)
        
        #Grab EndDate
        EndDate = str(record['008'])[17:21]
        EndDate = str(EndDate)
        print(EndDate)
        
        #Grab Language
        Language = str(record['008'])[41:44]
        ELanguage = str(Language)
        print(Language)
        
        #Grab Tag Number
        if record['099'] is None:
            print("No Tag Number")
        else:
            TagNumber = str(record['099']['a'])
            print(TagNumber)
            
        #Grab Creator
        if record['100'] is None:
            print("No Creator")
        else:
            Creator = str(record.author())
            print(Creator)
            
        #Grab Title
        if record['245'] is None:
            print("No Title")
        else:
            Title = str(record.title())
            print(Title)
            
        #Grab Alternative titles
        if record['246'] is None:
            print("No Alternative Titles")
        else:
            for field in record.get_fields('246'):
                AlternativeTitles = str(field)[10:] + " | " + AlternativeTitles 
            print(AlternativeTitles)
        
        #Grab Source of Title
        if record['588'] is None:
            print("No Source of Title")
        else:
            SourceOfTitle = str(record['588']['a'])
            print(SourceOfTitle)      
            
        #Grab PlaceManufacture
        if record['264'] is None:
            print("No Place of Manufacture")
        else:
            PlaceManufacture = str(record['264']['a'][:-1])
            print(PlaceManufacture)  
            
        #Grab NameManufacturer
        if record['264'] is None:
            print("No Name of Manufacture")
        else:
            NameManufacturer = str(record['264']['b'][:-1])
            print(NameManufacturer) 
            
        #Grab DateManufacurer
        if record['264'] is None:
            print("No Date of Manufacture")
        else:
            DateManufacturer = str(record['264']['c'])
            print(DateManufacturer) 
            
        #Grab ObjectExtent
        if record['300'] is None:
            print("No Object Extent")
        else:
            ObjectExtent = str(record['300']['a'])
            print(ObjectExtent) 
            
        #Grab Other Physical Details
        if record['300'] is None:
            print("No Material Type")
        else:
            ObjectMaterials = str(record['300']['b'])
            print(ObjectMaterials)
            
        #Grab Height
        #NEED TO TEST FOR LENGTH ZERO
        if record['300'] is None:
            print("No height")
        else: 
            Dimensionstring = str(record['300']['c'])
            Height = re.findall("^\d+", Dimensionstring)[0]
            print('Height is: '+ Height)
            
        #Grab Width
        #NEED TO TEST FOR LENGTH ZERO
        if record['300'] is None:
            print("No width")
        else: 
            Dimensionstring = str(record['300']['c'])
            Width1 = re.findall("\ x \d+\ ", Dimensionstring)[0]
            Width2 = re.findall("\d+", Width1)[0]
            Width = Width2
            print('Width is: '+ Width)
            
        #Grab Length
        #NEED TO TEST FOR LENGTH ZERO
        if record['300'] is None:
            print("No length")
        else: 
            Dimensionstring = str(record['300']['c'])
            Length1 = re.findall("\ \d+\ c", Dimensionstring)[0]
            Length2 = re.findall("\d+", Length1)[0]
            Length = Length2
            print('Length is: '+ Length)
            
        #Grab AccompanyingMaterial
        if record['300'] is None:
            print("No Accompanying Material")
        else:
            AccompanyingMaterial = str(record['300']['e'])
            print(AccompanyingMaterial)
            
        #Grab Contents = ''
        if record['505'] is None:
            print("No Contents")
        else:
            Contents = str(record['505']['a'])
            print(Contents)
            
        #Grab RestrictionsMaterial = ''
        if record['506'] is None:
            print("No RestrictionsMaterial")
        else:
            RestrictionsMaterial = str(record['506']['a'])
            print(RestrictionsMaterial)
            
        #Grab BibCitation = ''
        if record['510'] is None:
            print("No BibCitation")
        else:
            BibCitation = str(record['510']['a'])
            print(BibCitation)
            
        #Grab BibCitationLoc = ''
        if record['510'] is None:
            print("No BibCitationLoc")
        else:
            BibCitationLoc = str(record['510']['c'])
            print(BibCitationLoc)
            
        #Grab Description = ''
        if record['520'] is None:
            print("No Description")
        else:
            Description = str(record['520']['a'])
            print(Description)
         
        #Grab Condition = ''
        if record['866'] is None:
            print("No Condition")
        else:
            Condition = str(record['866']['x'])
            print(Condition)
            
        #Grab GiftNotes
        if record['866'] is None:
            print("No GiftNotes")
        else:
            GiftNotes = str(record['866']['z'])
            print(GiftNotes)
            
        #Grab StorageNeeds
        if record['981'] is None:
            print("No StorageNeeds")
        else:
            StorageNeeds = str(record['981']['x'])
            print(StorageNeeds)
            
        #Grab Mesh1
        if record['650'] is None:
            print("No Subjects")
        else:
            Mesh1 = str(record['650']['a'])
            print(Mesh1)
            
        #Grab Mesh1Q
        if record['650'] is None:
            print("No Subject Qualifiers")
        else:
            Mesh1Q = str(record['650']['x'])
            print(Mesh1Q)
            
        #Grab Mesh2
        #skipping for now
        
        #Grab Mesh2Q
        #skipping for now
        
        #Grab ContributorInd
        if record['700'] is None:
            print("No ContributorInd")
        else:
            for field in record.get_fields('700'):
                ContributorInd = str(field)[10:] + " | " + ContributorInd
                print(ContributorInd)     
        
        #Grab ContributorCorp
        if record ['710'] is None:
            print("no corpcontrib")
            print("no collection")
        else:
            #has 710
            if record ['710']['5'] is None:
                ContributorCorp = str(record['710'])[10:] + ' | ' + ContributorCorp
            #has 710 with $5
            else:
                for field in record.get_fields('710'):
                    print(field)
                    if str(field) == '=710  2\$aPharmaceutical Collection.$5MnU': 
                    #Grab collection
                        Collection = str(field)[10:]
                    elif str(field) == '=710  2\$aRamsey County Medical Society Collection.$5MnU':
                    #Grab collection
                        Collection = str(field)[10:]
                    elif str(field) == '=710  2\$aRosenthal Infant Feeding Collection.$5MnU':
                    #Grab collection
                        Collection = str(field)[10:]
                        
            print("corpcontrib is: " + ContributorCorp)
            print("collection is:" + Collection)
                
        #Grab Location = ''
        if record['852'] is None:
            print("No Location")
        else:
            Location = str(record['852']['c'])
            print(Location)
            
        #Grab Barcode
        #skipping for now
        
        #Grab Inventory Number
        #skipping for now
        
        list_row = {"InputDate": InputDate, "Action": Action, "MMSID": MMSID, "DateType": DateType, "StartDate": StartDate,
            "EndDate": EndDate, "Language": Language, "TagNumber": TagNumber, "Creator": Creator, "Title": Title, 
            "AlternativeTitles": AlternativeTitles, "SourceOfTitle": SourceOfTitle, "PlaceManufacture": PlaceManufacture,
            "NameManufacturer": NameManufacturer, "DateManufacurer": DateManufacurer, "ObjectExtent": ObjectExtent,
            "ObjectMaterials": ObjectMaterials, "Height": Height, "Width": Width, "Length": Length, 
            "AccompanyingMaterial": AccompanyingMaterial, "Contents": Contents, "RestrictionsMaterial":RestrictionsMaterial,
            "BibCitation": BibCitation, "BibCitationLoc": BibCitationLoc, "Description": Description, "Condition": Condition,
            "GiftNotes": GiftNotes, "StorageNeeds": StorageNeeds, "Mesh1": Mesh1, "Mesh1Q": Mesh1Q, "Mesh2": Mesh2, 
            "Mesh2Q": Mesh2Q, "Collection": Collection, "ContributorInd": ContributorInd, "ContributorCorp": ContributorCorp,
            "Location": Location, "Barcode": Barcode, "InventoryNo": InventoryNo}
        
        dfmarc = dfmarc.append(list_row, ignore_index=True)
```
